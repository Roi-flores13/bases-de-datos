# Intro KNN

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=3, cluster_std=5)
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
x1 = X[:,0]
x2 = X[:,1]
```

```python
plt.scatter(x1, x2, c= y == 0, label = "Class 0")
plt.scatter(x1,x2, c= y == 1, label = "Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Data")
plt.legend()
plt.grid()
```

# Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier()
clf.fit(X,y)
clf.score(X,y)
```

# Train, test, split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test =train_test_split(X,y, test_size=0.2, random_state=3)
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
