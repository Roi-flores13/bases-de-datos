# Correlations made simple

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_diabetes(as_frame=True)
data.keys()
```

```python
print(data.DESCR)
```

```python
df = data.frame
df
```

```python
corr = df.corr()
```

```python
sns.heatmap(corr, annot=True)
```
