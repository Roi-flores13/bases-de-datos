```python
def calculo_porcentaje(row, col1, col2):
    if row[col2] != 0:
        return round((row[col2]/row[col1])*100, 1)
    else:
        return 0

df_rs["GS%"] = df_rs.apply(calculo_porcentaje, col1="G", col2="GS", axis=1)

id_min_perc = df_rs["GS%"].idxmin() 
id_max_perc = df_rs["GS%"].idxmax()
player_min_perc = df_rs["Player"][id_min_perc]
player_max_perc = df_rs["Player"][id_max_perc]
min_perc = df_rs["GS%"][id_min_perc]
max_perc =  df_rs["GS%"][id_max_perc]

print(f"El jugador que menos partidos empezó en la temporada fue {player_min_perc} con un porcentaje de {min_perc}% de partidos iniciados")
print(f"El jugador que mas partidos empezó en la temporada fue {player_max_perc} con un porcentaje de {max_perc}% de partidos iniciados")

        
```
